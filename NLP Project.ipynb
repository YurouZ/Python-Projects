```python
from nltk import tokenize
def gettextnltk(x):
    page = []
    page_content = x.strip().split('\n\n')
    for a in page_content:
        line = tokenize.sent_tokenize(a)
        para = []
        for i in line:
            para.append(i)
        page.append(para)
    return page
```

```python
def readFile(file):
    pdfFileObj = open(file, 'rb')
    pdfReader = slate.PDF(pdfFileObj)
    num = len(pdfReader)
    j = 0
    page = []
    while j < num:
        p1 = pdfReader[j]
        p2 = gettextnltk(p1)
        page.append(p2)
        j = j + 1
    return page
```

```python
def term_vecor(sentence):
    word_tokens = re.split(r'\W+',sentence)
    filtered_sentence = [] 
    for w in word_tokens: 
        if w not in stop_words:
            if w not in filtered_sentence:
                filtered_sentence.append(w) 
    return filtered_sentence
```

```python
t= []
for i in range(len(sentence)):
    t.append(term_vecor(sentence[i]))
    
```

```python
pos_tagger = 'C:\\Users\\xzyr3\\Desktop\\Data Science using Python\\stanford-postagger-2018-10-16\\stanford-postagger-2018-10-16\\models\\english-bidirectional-distsim.tagger'
jar = 'C:\\Users\\xzyr3\\Desktop\\Data Science using Python\\stanford-postagger-2018-10-16\\stanford-postagger-2018-10-16\\stanford-postagger-3.9.2.jar'
st = StanfordPOSTagger(pos_tagger,jar)
postag = []

for n in t1:
    pos = st.tag(n)
    postag.append(pos)

print(postag)  
```

```python
# long phrases
entity = []
index2 = []
for i in range(len(postag)):
    lphrase=[]
    length = []
    for j in range(len(postag[i])):
        phrase = []
        if postag[i][j][1] ==  'NNP':
            phrase.append(postag[i][j])            
            for m in range(j+1,len(postag[i])):
                if postag[i][m][1] ==  'NNP' or postag[i][m][1] ==  'CD':
                    phrase.append(postag[i][m])
                else:
                    break
        if len(phrase)>= 2:
            lphrase.append(phrase)
            a = len(phrase)
            length.append(a)
            n = length.index(max(length))
    if lphrase!=[]:
        entity.append(lphrase[n])
        index2.append(i)
print(entity)
print(index2)
```

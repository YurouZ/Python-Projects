```python
import PyPDF2 as pypd 
# pyPDF2 will help you to read texts from the PDF files, but 
# it will maintain the structure of the tables.
#
from nltk import tokenize
def gettextnltk(x):
    page = []
    page_content = x.extractText().strip().split('\n\n\n\n')
    for a in page_content:
        line = tokenize.sent_tokenize(a)
        para = []
        for i in line:
            para.append(i)
        page.append(para)
    return page
```

```python
def readFile(file):
    pdfFileObj = open(file, 'rb')
    pdfReader = pypd.PdfFileReader(pdfFileObj)
    num = pdfReader.numPages
    j = 0
    page = []
    while j < num:
        p1 = pdfReader.getPage(j)
        p2 = gettextnltk(p1)
        page.append(p2)
        j = j + 1
    return page
import os
files = []
dird = '/Users/wesley/Desktop/NLPASS2'
for r, d, f in os.walk(dird):
    for file in f:
        if '.pdf' in file:
            files.append(os.path.join(r, file))
files
```

## 3. Run the skip-gram code from the Jupyter Notebook discussed in class and in 5 sentences describe its role in finding term collocations.

```python
from collections import Counter
import itertools

import nltk
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
from scipy import sparse
from scipy.sparse import linalg 
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity  
```

```python
def tokenizer(texts):
    stopwords_set = set(stopwords.words('english'))
    token = []

    for text in texts :
    
        tok = text.split()
        for t in tok:
            if t not in stopwords_set:
                token.append(t)
    return token 
```

```python
doc_token = []
for t in p:
    for x in t:
        doc_token.append(tokenizer(x))
doc_token   
```

```python
tok2indx = dict()
unigram_counts = Counter()
for ii, h in enumerate(doc_token):
    for token in h:
        unigram_counts[token] += 1    

    if token not in tok2indx:
        tok2indx[token] = len(tok2indx)
indx2tok = {indx:tok for tok,indx in tok2indx.items()}        
tok2indx
```

```python
print('vocabulary size: {}'.format(len(unigram_counts)))
print('most common: {}'.format(unigram_counts.most_common(100)))
```

### The role of skip-gram code in finding term collocations
### Go through every word of every page in the document, count the numbers for every group consisted of two words close to each other (next to each other or with only one interval).
### Once finish 10% pages further, print a notice sentence.
### In the end, print how many the different kinds of two-word-groups is and print 10 of them with the highest frequency in the document.

```python
back_window = 2
front_window = 2
skipgram_counts = Counter()
for ih, h in enumerate(doc_token):
    for ifw, fw in enumerate(h):
        icw_min = max(0, ifw - back_window)
        icw_max = min(len(h) - 1, ifw + front_window)
        icws = [ii for ii in range(icw_min, icw_max + 1) if ii != ifw]
        for icw in icws:
            skipgram = (h[ifw], h[icw])
            skipgram_counts[skipgram] += 1    
    if ih % 10 == 0:
        print(f'finished {ih/len(doc_token):.2%} of headlines')
        

print('number of skipgrams: {}'.format(len(skipgram_counts)))
print('most common: {}'.format(skipgram_counts.most_common(10)))
```
